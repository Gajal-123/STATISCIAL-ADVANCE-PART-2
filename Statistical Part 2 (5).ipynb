1.  What is hypothesis testing in statistics?
    -
    Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It helps determine whether the evidence from the data is strong enough to support or reject a claim (called a hypothesis) about the population.

 Key Concepts in Hypothesis Testing:
 
 1.Null Hypothesis (H0)
 - The default or original claim.

 - Usually states ‚Äúno effect‚Äù or ‚Äúno difference‚Äù.

 - Example: H0: The mean exam score is equal to 70.

 2.Alternative Hypothesis (H1 or Ha)
 - The competing claim you're trying to prove.

 - It states there is an effect, change, or difference.

 - Example: Ha: The mean exam score is not equal to 70.

 3.Test Statistic
 - A numerical value calculated from sample data, used to test the hypothesis.

 - Example: Z-score, t-score, chi-square, etc.

 4.Significance Level (Œ±)
 - The probability of rejecting the null hypothesis when it's actually true (Type I error).

 - Common values: 0.05, 0.01, or 0.10.

 5.p-value
 - The probability of obtaining a test statistic as extreme as the one observed, assuming H0is true.

 - If p-value ‚â§ Œ±, reject H0 ; otherwise, fail to reject H0

#####  Steps in Hypothesis Testing:

| Step | Description                                     |
| ---- | ----------------------------------------------- |
| 1.   | State the null and alternative hypotheses       |
| 2.   | Choose a significance level ($\alpha$)          |
| 3.   | Collect data and calculate the test statistic   |
| 4.   | Find the p-value or critical value              |
| 5.   | Make a decision: reject or fail to reject $H_0$ |
| 6.   | State the conclusion in context                 |

-> Example:

You want to test whether a new drug improves recovery time compared to the standard drug.
 - H0: New drug has no effect on recovery time.
 - Ha: New drug reduces recovery time

After analysis:

 - You find a p-value = 0.03
 - Significance level = 0.05

Since 0.03<0.05, you reject H0 ‚Üí There is significant evidence that the new drug helps.


| Term                               | Meaning                                    |
| ---------------------------------- | ------------------------------------------ |
| **Hypothesis Testing**             | Decision-making process using sample data  |
| **Null Hypothesis** ($H_0$)        | No effect or no difference                 |
| **Alternative Hypothesis** ($H_a$) | What you‚Äôre trying to prove                |
| **p-value**                        | Probability of observed result under $H_0$ |
| **Conclusion**                     | Reject or fail to reject $H_0$             |


2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?
    -
    In hypothesis testing, the null hypothesis and the alternative hypothesis are two competing statements about a population parameter. These form the basis for making statistical decisions using sample data.

Null Hypothesis (H0):
 - It represents the default, status quo, or no-effect assumption.

 - It assumes that any observed differences or effects in the sample data are due to random chance.

 - The null hypothesis is the one we test against and attempt to disprove.

-> Examples:
 - H0 : The population mean is equal to 50.
 - H0 : There is no difference between two treatments.
 - H0 : The coin is fair (P=0.5).


Alternative Hypothesis (Ha or H1 ):
 - This is the opposite of the null hypothesis.

 - It represents the statement we want to test for or prove.

 - It suggests that the observed effect is real and significant, not just due to chance.

-> Examples:
 - Ha : The population mean is not equal to 50.
 - Ha : There is a difference between two treatments.
 - Ha : The coin is biased (ùëÉ‚â†0.5).


##### Key Differences:

| Feature               | Null Hypothesis ($H_0$)                       | Alternative Hypothesis ($H_a$)            |
| --------------------- | --------------------------------------------- | ----------------------------------------- |
| **Purpose**           | Assumed true unless proven otherwise          | What you seek to provide evidence for     |
| **Represents**        | No change, no effect, status quo              | A new effect, difference, or relationship |
| **Decision Based On** | p-value vs. significance level ($\alpha$)     | Rejected if $p \leq \alpha$               |
| **Mathematical Form** | Usually includes **equality** (e.g., =, ‚â•, ‚â§) | Usually includes **inequality** (‚â†, <, >) |


##### Example: Testing a New Drug

| Hypothesis | Statement                                                    |
| ---------- | ------------------------------------------------------------ |
| $H_0$      | The new drug has **no effect** (mean recovery time = 5 days) |
| $H_a$      | The new drug **reduces recovery time** (mean < 5 days)       |

4.  What does a P-value represent in hypothesis testing?
    -
    The P-value is a key concept in hypothesis testing. It measures the probability of observing the sample results, or something more extreme, assuming the null hypothesis is true.

 - Definition:
The P-value is the probability of getting a result as extreme as (or more extreme than) the observed result by random chance alone, if the null hypothesis (H0) is true.

##### How to Interpret a P-value:
| **P-value**                      | **Interpretation**                                                       |
| -------------------------------- | ------------------------------------------------------------------------ |
| **Small P-value** (e.g., ‚â§ 0.05) | **Strong evidence** against $H_0$ ‚Üí **Reject** the null hypothesis       |
| **Large P-value** (e.g., > 0.05) | **Weak evidence** against $H_0$ ‚Üí **Fail to reject** the null hypothesis |

Example:

You test whether a coin is fair.
 - H0: The coin is fair (P=0.5)
 - You flip it 10 times and get 9 heads.

You calculate a P-value = 0.02

With a significance level Œ± = 0.05:

 - Since 0.02 < 0.05, you reject H0.
 - Conclusion: There's strong evidence the coin is not fair.


##### Important Notes:
 - A small P-value means your data is unlikely under the null hypothesis.

 - A large P-value doesn‚Äôt prove H0 is true ‚Äî it just means there's not enough evidence to reject it.

 - The P-value does not give the probability that H0 is true or false.


##### In Summary:

| Term             | Meaning                                                                         |
| ---------------- | ------------------------------------------------------------------------------- |
| **P-value**      | Probability of the observed result (or more extreme) occurring if $H_0$ is true |
| **Used for**     | Comparing with significance level (Œ±) to decide whether to reject $H_0$         |
| **Low P-value**  | Strong evidence against $H_0$                                                   |
| **High P-value** | Weak evidence; do not reject $H_0$                                              |

6.  What are Type 1 and Type 2 errors in hypothesis testing?
    -
    In hypothesis testing, Type I and Type II errors represent two kinds of incorrect conclusions that can occur when making decisions based on sample data.


##### Type I Error (False Positive):

 - Definition: Rejecting the null hypothesis (H0) when it is actually true.

 - It means you conclude there is an effect or difference when there isn‚Äôt.

 - Denoted by: Œ± (alpha) ‚Äî the significance level.

 - Example: Saying a drug works when it actually doesn't.


##### Type II Error (False Negative):

 - Definition: Failing to reject the null hypothesis (H0) when it is actually false.

 - It means you conclude there is no effect, when there actually is one.

 - Denoted by: Œ≤ (beta) ‚Äî the probability of Type II error.

 - Example: Saying a drug doesn't work when it actually does.


##### Comparison Table:\
| **Error Type**  | **What happens**                     | **Truth**               | **Decision**   | **Consequence**                      |
| --------------- | ------------------------------------ | ----------------------- | -------------- | ------------------------------------ |
| **Type I (Œ±)**  | Reject $H_0$ when it‚Äôs true          | $H_0$ is actually true  | False positive | You detect an effect that isn‚Äôt real |
| **Type II (Œ≤)** | Fail to reject $H_0$ when it‚Äôs false | $H_0$ is actually false | False negative | You miss detecting a real effect     |


##### Visualization Summary:
|                          | **Reality: $H_0$ is True** | **Reality: $H_0$ is False** |
| ------------------------ | -------------------------- | --------------------------- |
| **Reject $H_0$**         | **Type I Error (Œ±)**       | ‚úÖ Correct Decision          |
| **Fail to Reject $H_0$** | ‚úÖ Correct Decision         | **Type II Error (Œ≤)**       |


-> Key Point:

You cannot eliminate both errors, so:

 - Lowering Œ± reduces Type I error risk, but may increase Type II error.
 - A good balance is important and depends on the context (e.g., medicine vs. marketing).


| Error       | Meaning                                            | Symbol | Goal                          |
| ----------- | -------------------------------------------------- | ------ | ----------------------------- |
| **Type I**  | False alarm (rejecting a true $H_0$)               | Œ±      | Keep it low (e.g., 0.05)      |
| **Type II** | Missed detection (failing to reject a false $H_0$) | Œ≤      | Minimize with good test power |

14.  What is a confidence interval, and how is it used to interpret statistical results?
     -
     A confidence interval (CI) is a range of values derived from sample data that's used to estimate an unknown population parameter‚Äîsuch as a mean, proportion, or difference between groups. Unlike a point estimate (e.g., ‚Äúaverage = 50‚Äù), a CI gives both an estimate and a sense of how much uncertainty surrounds it.

 - A point estimate (e.g., sample mean)
 - A margin of error (critical value √ó standard error)
 - A chosen confidence level (commonly 90%, 95%, or 99%)

##### How It‚Äôs Calculated

 1.Point estimate: the statistic from your sample (mean, proportion, etc.).

 2.Standard error (SE): measures variability of the estimate‚Äîe.g. sample SD divided by ‚àön .

 3.Critical value: from a Z or t distribution based on your confidence level (for 95% using z ‚âà‚ÄØ1.96)

 4.Construct CI:

       Point¬†estimate¬±(critical¬†value√óSE)

 - A 95% CI doesn‚Äôt mean there's a 95% probability that this interval contains the true value. The true parameter is fixed‚Äîeither it's in the interval or not. 

 - Instead, it means: if we repeated this study many times, about 95% of those intervals would include the true parameter.

Common intuitive summary: ‚ÄúWe‚Äôre 95% confident the true value lies between L and U.‚Äù

##### Why It's Useful
 - Quantifies uncertainty: Wider CI ‚Üí less precision; narrower CI ‚Üí more precise estimate.

   - If it includes 0, it may not be significant. 

 - Gives range of plausible values: More informative than just ‚Äúsignificant/not significant.‚Äù


##### Key Points
 - Higher confidence levels (like 99%) yield wider intervals, trading precision for certainty.

 - The CI reflects the reliability of the estimation method, not the probability distribution of the true parameter. 


##### Best Practice Interpretation
‚Ä¢ When reporting:

‚ÄúAt the 95% confidence level, we estimate the [parameter] is between [L] and [U]. Because the interval excludes [null], the result is statistically significant.‚Äù
‚Ä¢ Use CI to understand both the strength of the effect and the uncertainty around it.


##### Common Misinterpretations
 - "There‚Äôs a 95% chance the true mean is in this interval." (Wrong; the interval is fixed once computed.)

 - "95% of the data lie within this interval." (Wrong; it's about the parameter, not data distribution.)

 - "This CI is a prediction for individual outcomes." (Wrong; that would be a prediction interval.)

20. What is an ANOVA test, and what are its assumptions?
    -
    An ANOVA (Analysis of Variance) is a statistical method used to test whether three or more group means are equal by comparing:

 - Between-group variance (differences in group means)

 - Within-group variance (natural variability inside each group)
Using the F-distribution, ANOVA yields an F-statistic assessing if the between-group variance is unusually large compared to within-group variance.


 1.When to Use ANOVA
 - One-way ANOVA: Compare means across k independent groups (k ‚â• 3).

 - Two-way ANOVA: Examine effects of two categorical factors and their interaction.

 - Extensions: Includes factorial designs, repeated-measures ANOVA, ANCOVA, mixed models.


 2.How ANOVA Works
 - Null hypothesis (H‚ÇÄ): All group population means are equal.

 - Calculate:
   - SS_between / df_between = mean square between
   - SS_within / df_within = mean square within
   - F-statistic = MS_between √∑ MS_within

 - Decision: If F is large (p-value < Œ±), reject H‚ÇÄ ‚Üí means differ.


 3.Key Assumptions of ANOVA
 
ANOVA relies on these critical assumptions:

 - Independence: Observations (and residuals) are independent; one observation doesn't affect another 

 - Normality: Residuals (or group-level data) should follow a normal distribution; moderate deviations are tolerated with larger or balanced samples 

 - Homoscedasticity: Variance across groups is equal. Testable via Levene‚Äôs or similar; ANOVA is robust if group sizes are equal and variance
   ratios < 3‚Äì4 

 - Measurement level: Dependent variable must be continuous (interval/ratio); independent variables are categorical.

 - (For repeated-measures) Sphericity: Variances of pairwise differences are equal; violations require corrections.


 4.When Assumptions Fail
 - Independence violations are critical‚Äîuse different designs instead (e.g., repeated measures).

 - Normality: With large, balanced groups, ANOVA remains robust; for severe departures, consider transformations or non-parametric alternatives.

 - Unequal variances: If groups are unbalanced or variance ratio > 4, use Welch‚Äôs ANOVA or adjust via robust methods.


 5.Post-hoc Testing
 - ANOVA only tells if there's a difference‚Äînot which groups differ. If significant, use post-hoc tests (Tukey HSD, Bonferroni, Scheff√©) to identify pairwise differences while controlling the Type I error rate.

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
    -
    To demonstrate the Central Limit Theorem (CLT) using Python, we can:

Draw samples from a non-normal distribution (e.g., exponential),

Take the means of many random samples from it,

Then visualize the distribution of these sample means, which will approximate a normal distribution.

```python
#Python Implementation of CLT using Exponential Distribution

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Parameters
population_size = 100000
sample_size = 30
num_samples = 1000

# Generate a non-normal (exponential) distribution
population = np.random.exponential(scale=2.0, size=population_size)

# Collect sample means
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Plot the original distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, kde=True, color='tomato', stat='density')
plt.title('Original Population (Exponential Distribution)')
plt.xlabel('Value')
plt.ylabel('Density')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='seagreen', stat='density')
plt.title(f'Distribution of Sample Means (n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

 - Sample means from this distribution approximate a normal distribution, even though the original is not.

15. Implement hypothesis testing using Z-statistics for a sample dataset.
    -
    Here's a Python implementation of hypothesis testing using Z-statistics, including:
 - A sample dataset,
 - Steps for a one-sample Z-test,
 - Interpretation of results.

 Scenario:
 - You're testing if the average weight of a product differs from a claimed weight of Œº‚ÇÄ = 50 kg.
 - You take a sample and want to test this claim at a 5% significance level (Œ± = 0.05).

```python
#Python Code: Z-Test Implementation

import numpy as np
from scipy.stats import norm

# Sample data (e.g., weights of items in kg)
sample_data = np.array([52, 49, 51, 50, 48, 53, 51, 50, 49, 52])

# Hypothesized population mean (Œº‚ÇÄ)
mu_0 = 50

# Population standard deviation (œÉ) - known
population_std = 2

# Sample mean and size
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Calculate Z-statistic
z_score = (sample_mean - mu_0) / (population_std / np.sqrt(n))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Significance level
alpha = 0.05

# Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z_score:.2f}")
print(f"P-value = {p_value:.4f}")

# Hypothesis testing decision
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference detected.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

 Hypothesis:
 - Null (H‚ÇÄ): Œº = 50 (no difference)
 - Alternative (H‚ÇÅ): Œº ‚â† 50 (there is a difference).


Assumptions:

16.  Create a confidence interval for a dataset using Python and interpret the result.
     -
     Here's a Python example that:
 - Calculates a confidence interval (CI) for a sample mean,
 - Assumes population standard deviation is unknown (so it uses a t-distribution),
 - Interprets the result.

```python
# Python Code: Confidence Interval for a Sample Mean

import numpy as np
from scipy import stats

# Sample dataset (e.g., test scores, weights, etc.)
data = np.array([52, 49, 51, 50, 48, 53, 51, 50, 49, 52])

# Sample statistics
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data)  # Standard Error of the Mean

# Confidence level (e.g., 95%)
confidence = 0.95

# Calculate t critical value
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

# Margin of error
margin_of_error = t_critical * std_err

# Confidence Interval
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

# Output
print(f"Sample Mean = {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

I|nterpretation:
"We are 95% confident that the true population mean lies between CI_lower and CI_upper."

This does not mean there's a 95% probability the mean is in this interval. Instead, if we repeat this sampling process many times, about 95% of the intervals we compute would contain the true mean.

5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
    -
    Great! Here's a Python function that:

Simulates hypothesis testing for a normal distribution,

Calculates Type I error (Œ±) and Type II error (Œ≤),

Visualizes both errors with the decision boundary.

```python
#Python Code: Type I and Type II Error Visualization

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0=100, mu_1=105, sigma=10, n=30, alpha=0.05):
    """
    mu_0: population mean under null hypothesis
    mu_1: actual mean under alternative hypothesis
    sigma: population standard deviation
    n: sample size
    alpha: significance level (Type I error rate)
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for a one-tailed test (right-tailed)
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu_0 + z_crit * se

    # Type I Error region (alpha)
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    y_null = norm.pdf(x, mu_0, se)
    y_alt = norm.pdf(x, mu_1, se)

    # Type II error: P(failing to reject H0 when H1 is true)
    beta = norm.cdf((x_crit - mu_1) / se)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label='H‚ÇÄ: Œº = {}'.format(mu_0), color='blue')
    plt.plot(x, y_alt, label='H‚ÇÅ: Œº = {}'.format(mu_1), color='green')

    # Shade Type I error (right tail of H0)
    plt.fill_between(x, y_null, 0, where=(x >= x_crit), color='red', alpha=0.4, label='Type I Error (Œ±)')

    # Shade Type II error (left of x_crit under H1)
    plt.fill_between(x, y_alt, 0, where=(x < x_crit), color='orange', alpha=0.4, label='Type II Error (Œ≤)')

    plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Critical Value: {x_crit:.2f}")
    print(f"Type I Error (Œ±): {alpha}")
    print(f"Type II Error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1 - Œ≤): {1 - beta:.4f}")

# Run the function
visualize_type1_type2_errors()
```

 Explanation:
 - H‚ÇÄ (null): The assumed population mean (mu_0)
 - H‚ÇÅ (alternative): The actual/true mean (mu_1)
 - Type I Error (Œ±): Rejecting H‚ÇÄ when it‚Äôs true ‚Äî red region
 - Type II Error (Œ≤): Failing to reject H‚ÇÄ when H‚ÇÅ is true ‚Äî orange region
 - Power: Probability of correctly rejecting H‚ÇÄ (1 ‚àí Œ≤)

9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
    -
    Python Function: Confidence Interval for Sample Mean

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate confidence interval for a sample mean.
    
    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (default = 0.95 for 95%)

    Returns:
        Tuple: (lower bound, upper bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return (lower_bound, upper_bound), mean, margin_of_error

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

(ci_low, ci_high), mean, margin = confidence_interval(sample_data)
print(f"Sample Mean          = {mean:.2f}")
print(f"95% Confidence Interval = ({ci_low:.2f}, {ci_high:.2f})")
print(f"Margin of Error      = ¬±{margin:.2f}")
```

Significance of a Confidence Interval:
 - A confidence interval gives a range of values within which the true population mean is likely to fall.
 - A 95% CI means that if we took 100 different samples and built a CI from each, about 95 of them would contain the true population mean.
 - It provides both:
   - Estimate: Sample mean
   - Uncertainty: Margin of error

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
    -
    Python Program: Margin of Error Calculation

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.
    
    Parameters:
        data (list or np.array): Sample data
        confidence (float): Confidence level (default = 0.95 for 95%)

    Returns:
        float: Margin of error
    """
    n = len(data)
    std_dev = np.std(data, ddof=1)
    std_err = std_dev / np.sqrt(n)

    # Get t-critical value
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)
    
    # Margin of error
    moe = t_crit * std_err
    return moe

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=25)

moe = calculate_margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): ¬±{moe:.2f}")
```

Explanation:
 - Margin of Error (MOE) quantifies the range within which the true population mean is likely to lie.

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
    -
    Bayes‚Äô Theorem

                  P(H‚à£D)= P(D‚à£H)‚ãÖP(H)/P(D)

Where:

P(H‚à£D): Posterior (probability of hypothesis H given data D)

P(H): Prior (belief about H before seeing data)

P(D‚à£H): Likelihood (probability of data under hypothesis)

P(D): Marginal likelihood (normalizing constant)


Scenario:
 - You take a test for a rare disease:
 - Disease prevalence (prior) = 1%
 - Test sensitivity (true positive rate) = 99%
 - Test specificity (true negative rate) = 95%
 - You tested positive. What‚Äôs the probability you actually have the disease?

```python
def bayes_theorem(prior, sensitivity, specificity):
    """
    Calculates posterior probability using Bayes' Theorem.
    
    prior: P(Disease) - prior probability of having the disease
    sensitivity: P(Pos|Disease) - true positive rate
    specificity: P(Neg|No Disease) - true negative rate
    """
    # P(Pos|No Disease)
    false_positive_rate = 1 - specificity
    
    # P(Disease|Pos) = [P(Pos|Disease) * P(Disease)] / P(Pos)
    numerator = sensitivity * prior
    denominator = (sensitivity * prior) + (false_positive_rate * (1 - prior))
    posterior = numerator / denominator
    return posterior

# Parameters
prior = 0.01           # 1% disease prevalence
sensitivity = 0.99     # 99% test sensitivity
specificity = 0.95     # 95% test specificity

# Compute posterior
posterior = bayes_theorem(prior, sensitivity, specificity)
print(f"Probability of having the disease after testing positive: {posterior:.4f} ({posterior*100:.2f}%)")
```

Explanation:
 - Prior: 1% of the population has the disease.
 - Likelihood: If you have the disease, the test detects it 99% of the time.
 - False Positive Rate: 5% of healthy people test positive.
 - Posterior: After testing positive, there's still only ~17% chance you actually have the disease ‚Äî due to the low prior and non-negligible false positives.

12. Perform a Chi-square test for independence between two categorical variables in Python.
    -
    Chi-Square Test for Independence in Python
    
Scenario:

You want to test whether gender and preference for a product are independent.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
# Example: Gender vs Product Preference
#              Like   Neutral  Dislike
# Male          20      15       10
# Female        30      25       15

data = np.array([[20, 15, 10],
                 [30, 25, 15]])

# Optional: convert to DataFrame for readability
df = pd.DataFrame(data, 
                  columns=['Like', 'Neutral', 'Dislike'],
                  index=['Male', 'Female'])

print("Contingency Table:")
print(df)

# Step 2: Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(data)

# Step 3: Output results
print("\nChi-Square Statistic:", round(chi2, 4))
print("Degrees of Freedom   :", dof)
print("P-Value              :", round(p, 4))
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, columns=df.columns, index=df.index))

# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("\n‚Üí Reject the null hypothesis: There is a significant association between gender and product preference.")
else:
    print("\n‚Üí Fail to reject the null hypothesis: No significant association found between gender and product preference.")
```

 Key Concepts:
 - Null Hypothesis (H‚ÇÄ): Variables are independent (no association).
 - Alternative Hypothesis (H‚ÇÅ): There is an association.
 - Use for analyzing two categorical variables (e.g., survey responses, gender vs outcome, etc.)

14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
     -
     Scenario:

You're rolling a 6-sided die 60 times and observing the outcomes. You want to test if the die is fair (i.e., all outcomes equally likely).

```python
#Python Code: Chi-square Goodness-of-Fit Test

import numpy as np
from scipy.stats import chisquare

# Step 1: Observed frequencies (from rolling a die 60 times)
observed = np.array([8, 9, 10, 12, 11, 10])  # total = 60

# Step 2: Expected frequencies for a fair die (uniform distribution)
expected = np.full_like(observed, fill_value=60/6)

# Step 3: Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Display results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"\nChi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("‚Üí Reject the null hypothesis: The die is not fair.")
else:
    print("‚Üí Fail to reject the null hypothesis: No evidence the die is unfair.")
```

 Goodness-of-Fit Test Recap:
 - Purpose: Test if observed data fits a theoretical distribution.
 - Null Hypothesis (H‚ÇÄ): Observed frequencies match the expected distribution.
 - Alternative Hypothesis (H‚ÇÅ): Observed frequencies differ significantly. 
 - Use: Coin/dice fairness, population proportions, genetics models, etc.
